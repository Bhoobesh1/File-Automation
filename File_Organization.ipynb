# Main Code For File Automation

```python
import os
def fileorganizer(path):
    extention=[]
    files=os.listdir(path)
    for file in files:
        if os.path.isfile(os.path.join(path,file)):
            x=os.path.splitext(file)[1][1:]
            if x not in extention:
                extention.append(x)
    print(extention)
    for folder in extention:
        try:
            x=os.path.join(path,folder)
            print(x)
            os.mkdir(x)
        except:
            pass
        for file in files:
            if os.path.isfile(os.path.join(path,file)):
                if(os.path.splitext(file)[1][1:]==os.path.split(x)[1]):
                    os.rename(os.path.join(path,file),os.path.join(x,file))
myfilepath=r"C:\Users\bhoob\OneDrive\Documents\fileorganizer"
fileorganizer(myfilepath)
```
